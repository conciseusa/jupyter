```python
%matplotlib widget
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
#%matplotlib ipympl

#%matplotlib inline
```

```python
# https://blogs.ed.ac.uk/mnaylor/2018/10/07/web-scraping-in-python/
# reading a log dir of a https://github.com/conciseusa/raspberry-pi-json-data-logger RPi on the local network
df = pd.read_csv('http://IP/log/2023-01-10-data.csv')
df
df.Time
```

```python
print(df.head())
```

```python
#plt.plot(df.Time, df.A3volt)
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111)
ax.plot(df.Time, df.A0tempF, label='A0 Temp')
ax.plot(df.Time, df.A2tempF, label='A2 Amb Temp')
ax.plot(df.Time, df.A3, label='A3 Power')
ax.plot(df.Time, df.duty_cycle_reducing, label='duty_cycle_reducing')
ax.plot(df.Time, df.A4volt, label='Battery V')
# ax.plot(df.Time, df.A2volt - df.A1volt, label='A2-A1 Battery A')
ax.set_xticks(np.arange(0, len(df.Time)+1, 60))
# ax.set_xticks(ax.get_xticks()[::200])
plt.grid()
plt.legend()
plt.gcf().autofmt_xdate()
plt.title("Arduino Data Display")
plt.show()
```
